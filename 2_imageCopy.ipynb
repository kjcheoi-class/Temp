```python
import cv2
from google.colab.patches import cv2_imshow

from google.colab import drive

# Google Drive 마운트
drive.mount('/content/drive')


# OpenCV 버전을 확인합니다
print("OpenCV version:")
print(cv2.__version__)

# 이미지를 읽어 옵니다.
img = cv2.imread("/content/drive/MyDrive/CV/Basic Vision Skills/rear_garden.PNG")

print("width: {} pixels".format(img.shape[1]))
print("height: {} pixels".format(img.shape[0]))
print("channels: {}".format(img.shape[2]))

# 이미지를 Display 합니다.
cv2_imshow(img)
```

```python
# 이미지를 복사합니다.
cv2.imwrite("/content/drive/MyDrive/CV/Basic Vision Skills/rear_garden_copy.PNG", img)
```
